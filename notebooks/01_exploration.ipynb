3. Identification des valeurs manquantes et aberrantes

3-1 Valeurs manquantes 

```python
missing_percent = df_chat.isnull().sum() * 100 / len(df_chat)

# Filtrage des colonnes avec au moins une valeur manquante
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)

# Affichage du pourcentage de chaque colonne
print(missing_percent)
```

```python
# Visualisation des valeurs manquantes

# Calculer le nombre de valeurs manquantes
missing_values = df_chat.isnull().sum()

# Filtrer les colonnes qui ont au moins une valeur manquante
missing_values = missing_values[missing_values > 0]

# Tracer un graphique
plt.figure(figsize=(10, 5))
missing_values.plot(kind="bar", color="red")
plt.xlabel("Colonnes")
plt.ylabel("Nombre de valeurs manquantes")
plt.title("Valeurs manquantes par colonne")
plt.xticks(rotation=45)
plt.show()
```
